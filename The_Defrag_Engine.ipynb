````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {}
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result):
        self.user = user_profile
        self.scan = daily_scan_result
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "text": f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into its Gift state today?",
            "tags": [theme, "Shadow Work", "Daily Log"]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {}, # NEW: Specific 3D Vector params
            "animation_states": []
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05, # Base rotation
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05 # Individual energy makes it spiky
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

if __name__ == "__main__":

    # 1. GENERATE PROFILES
    user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
    daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

    print(f"--- 1. DEFRAG SCAN (PHYSICS CORE) ---")
    defrag = DefragEngine(user_alpha, daily_transits)
    scan_result = defrag.run_daily_scan()
    print(json.dumps(scan_result, indent=2))

    print(f"\n--- 2. CURRICULUM & JOURNAL (STRATEGIC OUTPUT) ---")
    curriculum = CurriculumEngine(user_alpha, scan_result)
    print(json.dumps(curriculum.generate_curriculum(), indent=2))

    print(f"\n--- 3. CYMATIC RENDER & VECTOR CUBE (VISUAL OUTPUT) ---")
    cymatic = CymaticRenderer(user_alpha, daily_transits)
    print(json.dumps(cymatic.generate_config(), indent=2))
````

# Task
The task is to simulate an end-to-end Human Design system. This involves:

1.  **Simulating User Signup and Creating Primary Profile**: Defining natal details and creating a primary user profile.
2.  **Simulating Daily Transits**: Setting up mock daily transits.
3.  **Running DefragEngine**: Executing the DefragEngine with the primary profile and daily transits to get a daily scan result.
4.  **Running CurriculumEngine with Tone Adjustment**: Modifying the `CurriculumEngine` to accept a `tone_of_voice` parameter, then running it with the primary profile, daily scan, and a specified tone to generate journal prompts and lesson plans.
5.  **Running CymaticRenderer**: Executing the CymaticRenderer with the primary profile and daily transits to generate visual configurations.
6.  **Creating Secondary Profile for Relationship Analysis**: Generating a second distinct user profile.
7.  **Running RelationshipEngine**: Executing the RelationshipEngine with both primary and secondary profiles for relational dynamics.

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {}
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {}, # NEW: Specific 3D Vector params
            "animation_states": []
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05, # Base rotation
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05 # Individual energy makes it spiky
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

# Ensure user_alpha and scan_result are available from previous steps
# (These would be re-initialized if this cell was run in isolation)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}
defrag = DefragEngine(user_alpha, daily_transits)
scan_result = defrag.run_daily_scan()


print(f"\n--- 4. CURRICULUM & JOURNAL (STRATEGIC OUTPUT) with 'encouraging' tone ---")
curriculum_encouraging = CurriculumEngine(user_alpha, scan_result, tone_of_voice="encouraging")
print(json.dumps(curriculum_encouraging.generate_curriculum(), indent=2))

print(f"\n--- 5. CURRICULUM & JOURNAL (STRATEGIC OUTPUT) with 'scientific' tone ---")
curriculum_scientific = CurriculumEngine(user_alpha, scan_result, tone_of_voice="scientific")
print(json.dumps(curriculum_scientific.generate_curriculum(), indent=2))
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {}
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {} # Initialize for defrag visuals
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals() # Call the new method
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output: # Skip if no defrag output provided
            return

        # Shadow Work Visualization
        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate']) # Assuming sun_gate is the source of shadow
        shadow_gate_info = hexagrammatic_kernel['gates'].get(shadow_gate_str)

        self.config['defrag_visuals']['shadow_overlay_settings'] = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2 if shadow_theme != 'Unknown' else 0.0,
            "pulse_speed": 0.1 # Subtle pulse for internal processing
        }

        if shadow_gate_info:
            self.config['defrag_visuals']['focused_distortion'] = {
                "active_gate": int(shadow_gate_str),
                "distortion_intensity": 0.7, # Higher for active shadow
                "ripple_frequency": 0.5
            }
            self.config['defrag_visuals']['whispering_glyphs'] = {
                "active_gate": int(shadow_gate_str),
                "text": shadow_theme,
                "fade_duration": 3.0
            }

        # Opportunities Visualization
        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            self.config['defrag_visuals']['opportunities'].append({
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "ephemeral_text": opp['message']
            })



# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Updated CymaticRenderer (Defrag Visuals) ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
defrag_output = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
relationship_output = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
penta_output = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, defrag_output, tone_of_voice="encouraging")
curriculum_output = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=defrag_output,
    relationship_output=relationship_output,
    penta_output=penta_output,
    curriculum_output=curriculum_output
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with defrag visuals included. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {}
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {}, # Initialize for defrag visuals
            "relationship_visuals": {} # Initialize for relationship visuals
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals() # Call the new method
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output: # Skip if no defrag output provided
            return

        # Shadow Work Visualization
        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate']) # Assuming sun_gate is the source of shadow
        shadow_gate_info = hexagrammatic_kernel['gates'].get(shadow_gate_str)

        self.config['defrag_visuals']['shadow_overlay_settings'] = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2 if shadow_theme != 'Unknown' else 0.0,
            "pulse_speed": 0.1 # Subtle pulse for internal processing
        }

        if shadow_gate_info:
            self.config['defrag_visuals']['focused_distortion'] = {
                "active_gate": int(shadow_gate_str),
                "distortion_intensity": 0.7, # Higher for active shadow
                "ripple_frequency": 0.5
            }
            self.config['defrag_visuals']['whispering_glyphs'] = {
                "active_gate": int(shadow_gate_str),
                "text": shadow_theme,
                "fade_duration": 3.0
            }

        # Opportunities Visualization
        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            self.config['defrag_visuals']['opportunities'].append({
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "ephemeral_text": opp['message']
            })

    def _set_relationship_visuals(self):
        if not self.relationship_output: # Skip if no relationship output provided
            return

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            if spark['nature'] == "Harmonious":
                visual_props.update({
                    "color": "#ADD8E6", # Light blue / gold gradient
                    "animation_speed": 0.8, # Smooth, moderate
                    "particle_effect_type": "shimmer",
                    "glow_intensity": 1.2
                })
            elif spark['nature'] == "Volatile/Energizing":
                visual_props.update({
                    "color": "#FF4500", # Fiery orange-red / electric blue
                    "animation_speed": 2.0, # Faster, erratic
                    "particle_effect_type": "bursts",
                    "glow_intensity": 1.8
                })
            elif spark['nature'] == "Restrictive/Binding":
                visual_props.update({
                    "color": "#6A5ACD", # Dark grey / indigo
                    "animation_speed": 0.4, # Slow, deliberate
                    "particle_effect_type": "dense_mist",
                    "glow_intensity": 0.6
                })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        for friction in self.relationship_output.get('friction_points', []):
            # Placeholder for friction points visualization if they were present
            self.config['relationship_visuals']['friction_points'].append({
                "type": "conflict",
                "gates_involved": friction.get('gates', []), # Assuming friction output has 'gates'
                "distortion_intensity": 0.9,
                "color": "#FF0000", # Red for conflict
                "effect": "jagged_aura"
            })


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Updated CymaticRenderer (Defrag & Relationship Visuals) ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
defrag_output = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
relationship_output = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
penta_output = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, defrag_output, tone_of_voice="encouraging")
curriculum_output = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=defrag_output,
    relationship_output=relationship_output,
    penta_output=penta_output,
    curriculum_output=curriculum_output
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with defrag and relationship visuals included. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {}, # Initialize for defrag visuals
            "relationship_visuals": {}, # Initialize for relationship visuals
            "penta_visuals": {} # Initialize for penta visuals
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals() # Call the new method
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output: # Skip if no defrag output provided
            return

        # Shadow Work Visualization
        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate']) # Assuming sun_gate is the source of shadow
        shadow_gate_info = hexagrammatic_kernel['gates'].get(shadow_gate_str)

        self.config['defrag_visuals']['shadow_overlay_settings'] = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2 if shadow_theme != 'Unknown' else 0.0,
            "pulse_speed": 0.1 # Subtle pulse for internal processing
        }

        if shadow_gate_info:
            self.config['defrag_visuals']['focused_distortion'] = {
                "active_gate": int(shadow_gate_str),
                "distortion_intensity": 0.7, # Higher for active shadow
                "ripple_frequency": 0.5
            }
            self.config['defrag_visuals']['whispering_glyphs'] = {
                "active_gate": int(shadow_gate_str),
                "text": shadow_theme,
                "fade_duration": 3.0
            }

        # Opportunities Visualization
        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            self.config['defrag_visuals']['opportunities'].append({
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "ephemeral_text": opp['message']
            })

    def _set_relationship_visuals(self):
        if not self.relationship_output: # Skip if no relationship output provided
            return

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            if spark['nature'] == "Harmonious":
                visual_props.update({
                    "color": "#ADD8E6", # Light blue / gold gradient
                    "animation_speed": 0.8, # Smooth, moderate
                    "particle_effect_type": "shimmer",
                    "glow_intensity": 1.2
                })
            elif spark['nature'] == "Volatile/Energizing":
                visual_props.update({
                    "color": "#FF4500", # Fiery orange-red / electric blue
                    "animation_speed": 2.0, # Faster, erratic
                    "particle_effect_type": "bursts",
                    "glow_intensity": 1.8
                })
            elif spark['nature'] == "Restrictive/Binding":
                visual_props.update({
                    "color": "#6A5ACD", # Dark grey / indigo
                    "animation_speed": 0.4, # Slow, deliberate
                    "particle_effect_type": "dense_mist",
                    "glow_intensity": 0.6
                })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        for friction in self.relationship_output.get('friction_points', []):
            # Placeholder for friction points visualization if they were present
            self.config['relationship_visuals']['friction_points'].append({
                "type": "conflict",
                "gates_involved": friction.get('gates', []), # Assuming friction output has 'gates'
                "distortion_intensity": 0.9,
                "color": "#FF0000", # Red for conflict
                "effect": "jagged_aura"
            })

    def _set_penta_visuals(self):
        if not self.penta_output: # Skip if no penta output provided
            return

        # Functional Health Visualization
        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C", # Green for functional, Red for dysfunctional
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        # Circuitry Balance Visualization
        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        # Key Penta Channels Visualization
        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00", # Bright gold for active penta channels
                    "particle_type": "energy_flow"
                }
            })


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Updated CymaticRenderer (Defrag, Relationship & Penta Visuals) ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
defrag_output = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
relationship_output = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
penta_output = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, defrag_output, tone_of_voice="encouraging")
curriculum_output = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=defrag_output,
    relationship_output=relationship_output,
    penta_output=penta_output,
    curriculum_output=curriculum_output
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with defrag, relationship, and penta visuals included. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {},
            "relationship_visuals": {},
            "penta_visuals": {},
            "curriculum_visuals": {} # Initialize for curriculum visuals
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals() # Call the new method
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])
        shadow_gate_info = hexagrammatic_kernel['gates'].get(shadow_gate_str)

        self.config['defrag_visuals']['shadow_overlay_settings'] = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2 if shadow_theme != 'Unknown' else 0.0,
            "pulse_speed": 0.1
        }

        if shadow_gate_info:
            self.config['defrag_visuals']['focused_distortion'] = {
                "active_gate": int(shadow_gate_str),
                "distortion_intensity": 0.7,
                "ripple_frequency": 0.5
            }
            self.config['defrag_visuals']['whispering_glyphs'] = {
                "active_gate": int(shadow_gate_str),
                "text": shadow_theme,
                "fade_duration": 3.0
            }

        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            self.config['defrag_visuals']['opportunities'].append({
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700",
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "ephemeral_text": opp['message']
            })

    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            if spark['nature'] == "Harmonious":
                visual_props.update({
                    "color": "#ADD8E6",
                    "animation_speed": 0.8,
                    "particle_effect_type": "shimmer",
                    "glow_intensity": 1.2
                })
            elif spark['nature'] == "Volatile/Energizing":
                visual_props.update({
                    "color": "#FF4500",
                    "animation_speed": 2.0,
                    "particle_effect_type": "bursts",
                    "glow_intensity": 1.8
                })
            elif spark['nature'] == "Restrictive/Binding":
                visual_props.update({
                    "color": "#6A5ACD",
                    "animation_speed": 0.4,
                    "particle_effect_type": "dense_mist",
                    "glow_intensity": 0.6
                })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        for friction in self.relationship_output.get('friction_points', []):
            self.config['relationship_visuals']['friction_points'].append({
                "type": "conflict",
                "gates_involved": friction.get('gates', []),
                "distortion_intensity": 0.9,
                "color": "#FF0000",
                "effect": "jagged_aura"
            })

    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow"
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            # Theme-based visuals
            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            # Tone-of-voice-based visuals
            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            # Core Focus visuals
            if '31-7' in core_focus: # Example for a specific channel
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            # Action Step visuals
            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Updated CymaticRenderer (Defrag, Relationship, Penta & Curriculum Visuals) ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
defrag_output = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
relationship_output = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
penta_output = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, defrag_output, tone_of_voice="encouraging")
curriculum_output = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=defrag_output,
    relationship_output=relationship_output,
    penta_output=penta_output,
    curriculum_output=curriculum_output
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with defrag, relationship, penta, and curriculum visuals included. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {}, # Initialize for defrag visuals
            "relationship_visuals": {}, # Initialize for relationship visuals
            "penta_visuals": {}, # Initialize for penta visuals
            "curriculum_visuals": {} # Initialize for curriculum visuals
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])
        shadow_gate_info = hexagrammatic_kernel['gates'].get(shadow_gate_str)

        # Initialize shadow visual parameters with defaults
        shadow_overlay_settings = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2,
            "pulse_speed": 0.1,
            "overlay_texture": 'static_noise' if shadow_theme != 'Unknown' else 'none',
            "animation_curve": 'linear_fluctuation',
            "audio_cue": 'faint_static' if shadow_theme != 'Unknown' else 'none'
        }
        focused_distortion = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "distortion_intensity": 0.7,
            "ripple_frequency": 0.5,
            "distortion_waveform": 'perlin_noise',
            "audio_cue_intensity": 0.5
        }
        whispering_glyphs = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "text": shadow_theme,
            "fade_duration": 3.0,
            "glyph_particle_effect": 'dissolving_particles',
            "glyph_animation_type": 'floating'
        }

        # Retrieve tone of voice from curriculum output if available
        tone_of_voice = self.curriculum_output.get('recursive_journal_entry', {}).get('tone_of_voice', 'neutral')

        # Apply tone of voice adjustments to shadow visuals
        if tone_of_voice == 'encouraging':
            shadow_overlay_settings['desaturation_filter'] = max(0.0, shadow_overlay_settings['desaturation_filter'] - 0.1)
            focused_distortion['distortion_intensity'] = max(0.0, focused_distortion['distortion_intensity'] - 0.2)
            shadow_overlay_settings['audio_cue'] = 'subtle_chime'
            focused_distortion['audio_cue_intensity'] = max(0.0, focused_distortion['audio_cue_intensity'] - 0.2)
            whispering_glyphs['glyph_particle_effect'] = 'rising_sparkles'
            whispering_glyphs['glyph_animation_type'] = 'ascending'

        self.config['defrag_visuals']['shadow_overlay_settings'] = shadow_overlay_settings
        self.config['defrag_visuals']['focused_distortion'] = focused_distortion
        self.config['defrag_visuals']['whispering_glyphs'] = whispering_glyphs

        # Opportunities Visualization
        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            opportunity_visuals = {
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "particle_type": "sparkling_orbs", # New specific particle type
                "ephemeral_text": opp['message'],
                "channel_texture": 'flowing_energy',
                "channel_animation_curve": 'accelerating_flow',
                "audio_cue": 'chime'
            }

            # Apply tone of voice adjustments to opportunities
            if tone_of_voice == 'encouraging':
                opportunity_visuals['glow_intensity'] += 0.5 # Increase glow
                opportunity_visuals['glow_color'] = "#A3D4FF" # Brighter blue
                opportunity_visuals['channel_animation_curve'] = 'soaring_arc'
                opportunity_visuals['audio_cue'] = 'uplifting_swell'

            self.config['defrag_visuals']['opportunities'].append(opportunity_visuals)


    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            if spark['nature'] == "Harmonious":
                visual_props.update({
                    "color": "#ADD8E6",
                    "animation_speed": 0.8,
                    "particle_effect_type": "shimmer",
                    "glow_intensity": 1.2
                })
            elif spark['nature'] == "Volatile/Energizing":
                visual_props.update({
                    "color": "#FF4500",
                    "animation_speed": 2.0,
                    "particle_effect_type": "bursts",
                    "glow_intensity": 1.8
                })
            elif spark['nature'] == "Restrictive/Binding":
                visual_props.update({
                    "color": "#6A5ACD",
                    "animation_speed": 0.4,
                    "particle_effect_type": "dense_mist",
                    "glow_intensity": 0.6
                })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        for friction in self.relationship_output.get('friction_points', []):
            self.config['relationship_visuals']['friction_points'].append({
                "type": "conflict",
                "gates_involved": friction.get('gates', []),
                "distortion_intensity": 0.9,
                "color": "#FF0000",
                "effect": "jagged_aura"
            })

    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow"
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            # Theme-based visuals
            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            # Tone-of-voice-based visuals
            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            # Core Focus visuals
            if '31-7' in core_focus: # Example for a specific channel
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            # Action Step visuals
            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Updated CymaticRenderer (Defrag, Relationship, Penta & Curriculum Visuals) ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
du = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
ru = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
pu = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, du, tone_of_voice="encouraging")
cu = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=du,
    relationship_output=ru,
    penta_output=pu,
    curriculum_output=cu
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with defrag, relationship, penta, and curriculum visuals included. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {}, # Initialize for defrag visuals
            "relationship_visuals": {}, # Initialize for relationship visuals
            "penta_visuals": {},
            "curriculum_visuals": {} # Initialize for curriculum visuals
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])

        # Initialize shadow visual parameters with defaults
        shadow_overlay_settings = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2,
            "pulse_speed": 0.1,
            "overlay_texture": 'static_noise' if shadow_theme != 'Unknown' else 'none',
            "animation_curve": 'linear_fluctuation',
            "audio_cue": 'faint_static' if shadow_theme != 'Unknown' else 'none'
        }
        focused_distortion = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "distortion_intensity": 0.7,
            "ripple_frequency": 0.5,
            "distortion_waveform": 'perlin_noise',
            "audio_cue_intensity": 0.5
        }
        whispering_glyphs = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "text": shadow_theme,
            "fade_duration": 3.0,
            "glyph_particle_effect": 'dissolving_particles',
            "glyph_animation_type": 'floating'
        }

        # Retrieve tone of voice from curriculum output if available
        tone_of_voice = self.curriculum_output.get('recursive_journal_entry', {}).get('tone_of_voice', 'neutral')

        # Apply tone of voice adjustments to shadow visuals
        if tone_of_voice == 'encouraging':
            shadow_overlay_settings['desaturation_filter'] = max(0.0, shadow_overlay_settings['desaturation_filter'] - 0.1)
            focused_distortion['distortion_intensity'] = max(0.0, focused_distortion['distortion_intensity'] - 0.2)
            shadow_overlay_settings['audio_cue'] = 'subtle_chime'
            focused_distortion['audio_cue_intensity'] = max(0.0, focused_distortion['audio_cue_intensity'] - 0.2)
            whispering_glyphs['glyph_particle_effect'] = 'rising_sparkles'
            whispering_glyphs['glyph_animation_type'] = 'ascending'

        self.config['defrag_visuals']['shadow_overlay_settings'] = shadow_overlay_settings
        self.config['defrag_visuals']['focused_distortion'] = focused_distortion
        self.config['defrag_visuals']['whispering_glyphs'] = whispering_glyphs

        # Opportunities Visualization
        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            opportunity_visuals = {
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "particle_type": "sparkling_orbs", # New specific particle type
                "ephemeral_text": opp['message'],
                "channel_texture": 'flowing_energy',
                "channel_animation_curve": 'accelerating_flow',
                "audio_cue": 'chime'
            }

            # Apply tone of voice adjustments to opportunities
            if tone_of_voice == 'encouraging':
                opportunity_visuals['glow_intensity'] += 0.5 # Increase glow
                opportunity_visuals['glow_color'] = "#A3D4FF" # Brighter blue
                opportunity_visuals['channel_animation_curve'] = 'soaring_arc'
                opportunity_visuals['audio_cue'] = 'uplifting_swell'

            self.config['defrag_visuals']['opportunities'].append(opportunity_visuals)


    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            if spark['nature'] == "Harmonious":
                visual_props.update({
                    "color": "#ADD8E6",
                    "animation_speed": 0.8,
                    "particle_effect_type": "shimmer",
                    "glow_intensity": 1.2
                })
            elif spark['nature'] == "Volatile/Energizing":
                visual_props.update({
                    "color": "#FF4500",
                    "animation_speed": 2.0,
                    "particle_effect_type": "bursts",
                    "glow_intensity": 1.8
                })
            elif spark['nature'] == "Restrictive/Binding":
                visual_props.update({
                    "color": "#6A5ACD",
                    "animation_speed": 0.4,
                    "particle_effect_type": "dense_mist",
                    "glow_intensity": 0.6
                })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        for friction in self.relationship_output.get('friction_points', []):
            # Placeholder for friction points visualization if they were present
            self.config['relationship_visuals']['friction_points'].append({
                "type": "conflict",
                "gates_involved": friction.get('gates', []),
                "distortion_intensity": 0.9,
                "color": "#FF0000",
                "effect": "jagged_aura"
            })

    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow"
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            # Theme-based visuals
            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            # Tone-of-voice-based visuals
            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            # Core Focus visuals
            if '31-7' in core_focus: # Example for a specific channel
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            # Action Step visuals
            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Updated CymaticRenderer (Defrag, Relationship, Penta & Curriculum Visuals) ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
du = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
ru = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
pu = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, du, tone_of_voice="encouraging")
cu = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=du,
    relationship_output=ru,
    penta_output=pu,
    curriculum_output=cu
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with defrag, relationship, penta, and curriculum visuals included. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {}, # Initialize for defrag visuals
            "relationship_visuals": {}, # Initialize for relationship visuals
            "penta_visuals": {},
            "curriculum_visuals": {} # Initialize for curriculum visuals
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])

        # Initialize shadow visual parameters with defaults
        shadow_overlay_settings = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2,
            "pulse_speed": 0.1,
            "overlay_texture": 'static_noise' if shadow_theme != 'Unknown' else 'none',
            "animation_curve": 'linear_fluctuation',
            "audio_cue": 'faint_static' if shadow_theme != 'Unknown' else 'none'
        }
        focused_distortion = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "distortion_intensity": 0.7,
            "ripple_frequency": 0.5,
            "distortion_waveform": 'perlin_noise',
            "audio_cue_intensity": 0.5
        }
        whispering_glyphs = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "text": shadow_theme,
            "fade_duration": 3.0,
            "glyph_particle_effect": 'dissolving_particles',
            "glyph_animation_type": 'floating'
        }

        # Retrieve tone of voice from curriculum output if available
        tone_of_voice = self.curriculum_output.get('recursive_journal_entry', {}).get('tone_of_voice', 'neutral')

        # Apply tone of voice adjustments to shadow visuals
        if tone_of_voice == 'encouraging':
            shadow_overlay_settings['desaturation_filter'] = max(0.0, shadow_overlay_settings['desaturation_filter'] - 0.1)
            focused_distortion['distortion_intensity'] = max(0.0, focused_distortion['distortion_intensity'] - 0.2)
            shadow_overlay_settings['audio_cue'] = 'subtle_chime'
            focused_distortion['audio_cue_intensity'] = max(0.0, focused_distortion['audio_cue_intensity'] - 0.2)
            whispering_glyphs['glyph_particle_effect'] = 'rising_sparkles'
            whispering_glyphs['glyph_animation_type'] = 'ascending'

        self.config['defrag_visuals']['shadow_overlay_settings'] = shadow_overlay_settings
        self.config['defrag_visuals']['focused_distortion'] = focused_distortion
        self.config['defrag_visuals']['whispering_glyphs'] = whispering_glyphs

        # Opportunities Visualization
        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            opportunity_visuals = {
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "particle_type": "sparkling_orbs", # New specific particle type
                "ephemeral_text": opp['message'],
                "channel_texture": 'flowing_energy',
                "channel_animation_curve": 'accelerating_flow',
                "audio_cue": 'chime'
            }

            # Apply tone of voice adjustments to opportunities
            if tone_of_voice == 'encouraging':
                opportunity_visuals['glow_intensity'] += 0.5 # Increase glow
                opportunity_visuals['glow_color'] = "#A3D4FF" # Brighter blue
                opportunity_visuals['channel_animation_curve'] = 'soaring_arc'
                opportunity_visuals['audio_cue'] = 'uplifting_swell'

            self.config['defrag_visuals']['opportunities'].append(opportunity_visuals)


    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            if spark['nature'] == "Harmonious":
                visual_props.update({
                    "color": "#ADD8E6",
                    "animation_speed": 0.8,
                    "particle_effect_type": "shimmer",
                    "glow_intensity": 1.2
                })
            elif spark['nature'] == "Volatile/Energizing":
                visual_props.update({
                    "color": "#FF4500",
                    "animation_speed": 2.0,
                    "particle_effect_type": "bursts",
                    "glow_intensity": 1.8
                })
            elif spark['nature'] == "Restrictive/Binding":
                visual_props.update({
                    "color": "#6A5ACD",
                    "animation_speed": 0.4,
                    "particle_effect_type": "dense_mist",
                    "glow_intensity": 0.6
                })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        for friction in self.relationship_output.get('friction_points', []):
            # Placeholder for friction points visualization if they were present
            self.config['relationship_visuals']['friction_points'].append({
                "type": "conflict",
                "gates_involved": friction.get('gates', []),
                "distortion_intensity": 0.9,
                "color": "#FF0000",
                "effect": "jagged_aura"
            })

    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow"
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            # Theme-based visuals
            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            # Tone-of-voice-based visuals
            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            # Core Focus visuals
            if '31-7' in core_focus: # Example for a specific channel
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            # Action Step visuals
            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Updated CymaticRenderer (Defrag, Relationship, Penta & Curriculum Visuals) ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
du = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
ru = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
pu = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, du, tone_of_voice="encouraging")
cu = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=du,
    relationship_output=ru,
    penta_output=pu,
    curriculum_output=cu
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with defrag, relationship, penta, and curriculum visuals included. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {}, # Initialize for defrag visuals
            "relationship_visuals": {}, # Initialize for relationship visuals
            "penta_visuals": {},
            "curriculum_visuals": {} # Initialize for curriculum visuals
        }

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        # LOGIC: Define the 3D Vector Cube state based on Circuitry
        # X Axis = Logic (Collective), Y Axis = Tribal, Z Axis = Individual
        # Normalized vectors [0.0 to 1.0]

        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1 # Prevent div by zero

        vector = [
            counts['Collective'] / total, # X
            counts['Tribal'] / total,     # Y
            counts['Individual'] / total  # Z
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])

        # Initialize shadow visual parameters with defaults
        shadow_overlay_settings = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2,
            "pulse_speed": 0.1,
            "overlay_texture": 'static_noise' if shadow_theme != 'Unknown' else 'none',
            "animation_curve": 'linear_fluctuation',
            "audio_cue": 'faint_static' if shadow_theme != 'Unknown' else 'none'
        }
        focused_distortion = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "distortion_intensity": 0.7,
            "ripple_frequency": 0.5,
            "distortion_waveform": 'perlin_noise',
            "audio_cue_intensity": 0.5
        }
        whispering_glyphs = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "text": shadow_theme,
            "fade_duration": 3.0,
            "glyph_particle_effect": 'dissolving_particles',
            "glyph_animation_type": 'floating'
        }

        # Retrieve tone of voice from curriculum output if available
        tone_of_voice = self.curriculum_output.get('recursive_journal_entry', {}).get('tone_of_voice', 'neutral')

        # Apply tone of voice adjustments to shadow visuals
        if tone_of_voice == 'encouraging':
            shadow_overlay_settings['desaturation_filter'] = max(0.0, shadow_overlay_settings['desaturation_filter'] - 0.1)
            focused_distortion['distortion_intensity'] = max(0.0, focused_distortion['distortion_intensity'] - 0.2)
            shadow_overlay_settings['audio_cue'] = 'subtle_chime'
            focused_distortion['audio_cue_intensity'] = max(0.0, focused_distortion['audio_cue_intensity'] - 0.2)
            whispering_glyphs['glyph_particle_effect'] = 'rising_sparkles'
            whispering_glyphs['glyph_animation_type'] = 'ascending'

        self.config['defrag_visuals']['shadow_overlay_settings'] = shadow_overlay_settings
        self.config['defrag_visuals']['focused_distortion'] = focused_distortion
        self.config['defrag_visuals']['whispering_glyphs'] = whispering_glyphs

        # Opportunities Visualization
        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            opportunity_visuals = {
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "particle_type": "sparkling_orbs", # New specific particle type
                "ephemeral_text": opp['message'],
                "channel_texture": 'flowing_energy',
                "channel_animation_curve": 'accelerating_flow',
                "audio_cue": 'chime'
            }

            # Apply tone of voice adjustments to opportunities
            if tone_of_voice == 'encouraging':
                opportunity_visuals['glow_intensity'] += 0.5 # Increase glow
                opportunity_visuals['glow_color'] = "#A3D4FF" # Brighter blue
                opportunity_visuals['channel_animation_curve'] = 'soaring_arc'
                opportunity_visuals['audio_cue'] = 'uplifting_swell'

            self.config['defrag_visuals']['opportunities'].append(opportunity_visuals)


    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            if spark['nature'] == "Harmonious":
                visual_props.update({
                    "color": "#ADD8E6",
                    "animation_speed": 0.8,
                    "particle_effect_type": "shimmer",
                    "glow_intensity": 1.2
                })
            elif spark['nature'] == "Volatile/Energizing":
                visual_props.update({
                    "color": "#FF4500",
                    "animation_speed": 2.0,
                    "particle_effect_type": "bursts",
                    "glow_intensity": 1.8
                })
            elif spark['nature'] == "Restrictive/Binding":
                visual_props.update({
                    "color": "#6A5ACD",
                    "animation_speed": 0.4,
                    "particle_effect_type": "dense_mist",
                    "glow_intensity": 0.6
                })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        for friction in self.relationship_output.get('friction_points', []):
            # Placeholder for friction points visualization if they were present
            self.config['relationship_visuals']['friction_points'].append({
                "type": "conflict",
                "gates_involved": friction.get('gates', []),
                "distortion_intensity": 0.9,
                "color": "#FF0000",
                "effect": "jagged_aura"
            })

    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow"
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            # Theme-based visuals
            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            # Tone-of-voice-based visuals
            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            # Core Focus visuals
            if '31-7' in core_focus: # Example for a specific channel
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            # Action Step visuals
            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Updated CymaticRenderer (Defrag, Relationship, Penta & Curriculum Visuals) ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
du = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
ru = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
pu = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, du, tone_of_voice="encouraging")
cu = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=du,
    relationship_output=ru,
    penta_output=pu,
    curriculum_output=cu
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with defrag, relationship, penta, and curriculum visuals included. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {},
            "relationship_visuals": {},
            "penta_visuals": {},
            "curriculum_visuals": {}
        }

    def _is_user_pressured(self):
        # Helper to check if the main user (self.user) is under pressure from daily transits
        pressure_gates = [61, 63, 64]
        user_active_gates = self.user['mechanics']['active_gates'] # User's natal gates, not transit gates
        transit_active_gates = self.transits['mechanics']['active_gates'] # Active gates from today's transits

        # Check if the daily transits themselves are from a pressure gate
        if self.transits['mechanics']['sun_gate'] in pressure_gates:
            return True

        # Check if any of user's active gates are pressure gates and are part of the daily active transits (indirect pressure)
        if any(g in pressure_gates and g in transit_active_gates for g in user_active_gates):
            return True

        return False

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1

        vector = [
            counts['Collective'] / total,
            counts['Tribal'] / total,
            counts['Individual'] / total
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])

        shadow_overlay_settings = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2,
            "pulse_speed": 0.1,
            "overlay_texture": 'static_noise' if shadow_theme != 'Unknown' else 'none',
            "animation_curve": 'linear_fluctuation',
            "audio_cue": 'faint_static' if shadow_theme != 'Unknown' else 'none'
        }
        focused_distortion = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "distortion_intensity": 0.7,
            "ripple_frequency": 0.5,
            "distortion_waveform": 'perlin_noise',
            "audio_cue_intensity": 0.5
        }
        whispering_glyphs = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "text": shadow_theme,
            "fade_duration": 3.0,
            "glyph_particle_effect": 'dissolving_particles',
            "glyph_animation_type": 'floating'
        }

        tone_of_voice = self.curriculum_output.get('recursive_journal_entry', {}).get('tone_of_voice', 'neutral')

        if tone_of_voice == 'encouraging':
            shadow_overlay_settings['desaturation_filter'] = max(0.0, shadow_overlay_settings['desaturation_filter'] - 0.1)
            focused_distortion['distortion_intensity'] = max(0.0, focused_distortion['distortion_intensity'] - 0.2)
            shadow_overlay_settings['audio_cue'] = 'subtle_chime'
            focused_distortion['audio_cue_intensity'] = max(0.0, focused_distortion['audio_cue_intensity'] - 0.2)
            whispering_glyphs['glyph_particle_effect'] = 'rising_sparkles'
            whispering_glyphs['glyph_animation_type'] = 'ascending'

        self.config['defrag_visuals']['shadow_overlay_settings'] = shadow_overlay_settings
        self.config['defrag_visuals']['focused_distortion'] = focused_distortion
        self.config['defrag_visuals']['whispering_glyphs'] = whispering_glyphs

        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            opportunity_visuals = {
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700",
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "particle_type": "sparkling_orbs",
                "ephemeral_text": opp['message'],
                "channel_texture": 'flowing_energy',
                "channel_animation_curve": 'accelerating_flow',
                "audio_cue": 'chime'
            }

            if tone_of_voice == 'encouraging':
                opportunity_visuals['glow_intensity'] += 0.5
                opportunity_visuals['glow_color'] = "#A3D4FF"
                opportunity_visuals['channel_animation_curve'] = 'soaring_arc'
                opportunity_visuals['audio_cue'] = 'uplifting_swell'

            self.config['defrag_visuals']['opportunities'].append(opportunity_visuals)

    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        # Determine if the primary user (self.user) is pressured for influencing shared visuals
        user_is_pressured = self._is_user_pressured()

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            # Default values for new parameters
            geometry_deformation_type = 'none'
            geometry_deformation_intensity = 0.0
            particle_stream_direction = 'none'
            lighting_effect_type = 'none'
            environmental_aura_color = '#FFFFFF'
            audio_cue = 'none'

            if spark['nature'] == "Harmonious":
                geometry_deformation_type = 'subtle_pulse'
                geometry_deformation_intensity = 0.1
                particle_stream_direction = 'outward'
                lighting_effect_type = 'ambient_glow'
                environmental_aura_color = '#E0FFFF'
                audio_cue = 'harmonious_chime'
                visual_props.update({"color": "#ADD8E6", "animation_speed": 0.8, "particle_effect_type": "shimmer", "glow_intensity": 1.2})
            elif spark['nature'] == "Volatile/Energizing":
                geometry_deformation_type = 'erratic_shimmer'
                geometry_deformation_intensity = 0.4
                particle_stream_direction = 'random_flow'
                lighting_effect_type = 'focused_beam'
                environmental_aura_color = '#FFD700'
                audio_cue = 'static_hiss'
                visual_props.update({"color": "#FF4500", "animation_speed": 2.0, "particle_effect_type": "bursts", "glow_intensity": 1.8})
            elif spark['nature'] == "Restrictive/Binding":
                geometry_deformation_type = 'subtle_compression'
                geometry_deformation_intensity = 0.3
                particle_stream_direction = 'inward'
                lighting_effect_type = 'shadow_cast'
                environmental_aura_color = '#4B0082'
                audio_cue = 'deep_hum'
                visual_props.update({"color": "#6A5ACD", "animation_speed": 0.4, "particle_effect_type": "dense_mist", "glow_intensity": 0.6})

            # Apply influence from primary user's defrag output if they are source_1
            if user_is_pressured and spark['source_1_name'].startswith(self.user['name']):
                if 'Volatile' not in spark['nature']: # Only enhance non-volatile if user is pressured
                    geometry_deformation_intensity = min(1.0, geometry_deformation_intensity + 0.2)
                    particle_stream_direction = 'erratic_flicker'
                    lighting_effect_type = 'subtle_flicker'
                    audio_cue = 'underlying_tension'

            visual_props.update({
                "geometry_deformation_type": geometry_deformation_type,
                "geometry_deformation_intensity": geometry_deformation_intensity,
                "particle_stream_direction": particle_stream_direction,
                "lighting_effect_type": lighting_effect_type,
                "environmental_aura_color": environmental_aura_color,
                "audio_cue": audio_cue
            })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        # Mocking a friction point for demonstration if no actual friction points are generated
        if not self.relationship_output.get('friction_points'):
             # Example: create a mock friction point if source_1's sun gate is 38 (The Fighter - struggle)
            user_sun_gate = str(self.user['mechanics']['planetary_map'].get('Sun'))
            if user_sun_gate == '38' and hexagrammatic_kernel['gates'].get(user_sun_gate, {}).get('shadow') == 'Struggle':
                self.config['relationship_visuals']['friction_points'].append({
                    "type": "conflict",
                    "gates_involved": [38, hexagrammatic_kernel['harmonic_map'].get(38)], # Assuming a harmonic partner exists for potential conflict
                    "distortion_intensity": 0.9,
                    "color": "#FF0000",
                    "effect": "jagged_aura",
                    "geometry_deformation_type": 'fractured_surface',
                    "particle_stream_type": 'chaotic_bursts',
                    "lighting_effect_type": 'flickering_red',
                    "environmental_aura_color": '#FF8C00',
                    "audio_cue": 'dissonant_tone'
                })
        else:
            for friction in self.relationship_output.get('friction_points', []):
                # Existing friction points would get these parameters
                self.config['relationship_visuals']['friction_points'].append({
                    "type": "conflict",
                    "gates_involved": friction.get('gates', []),
                    "distortion_intensity": 0.9,
                    "color": "#FF0000",
                    "effect": "jagged_aura",
                    "geometry_deformation_type": 'fractured_surface',
                    "particle_stream_type": 'chaotic_bursts',
                    "lighting_effect_type": 'flickering_red',
                    "environmental_aura_color": '#FF8C00',
                    "audio_cue": 'dissonant_tone'
                })


    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow"
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            if '31-7' in core_focus:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Enhanced Relationship Visuals ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
du = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
ru = relationship_engine_instance.run_overlay()

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
pu = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, du, tone_of_voice="encouraging")
cu = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=du,
    relationship_output=ru,
    penta_output=pu,
    curriculum_output=cu
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with enhanced relationship visuals. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {},
            "relationship_visuals": {},
            "penta_visuals": {},
            "curriculum_visuals": {}
        }

    def _is_user_pressured(self):
        # Helper to check if the main user (self.user) is under pressure from daily transits
        pressure_gates = [61, 63, 64]
        user_active_gates = self.user['mechanics']['active_gates'] # User's natal gates, not transit gates
        transit_active_gates = self.transits['mechanics']['active_gates'] # Active gates from today's transits

        # Check if the daily transits themselves are from a pressure gate
        if self.transits['mechanics']['sun_gate'] in pressure_gates:
            return True

        # Check if any of user's active gates are pressure gates and are part of the daily active transits (indirect pressure)
        if any(g in pressure_gates and g in transit_active_gates for g in user_active_gates):
            return True

        return False

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1

        vector = [
            counts['Collective'] / total,
            counts['Tribal'] / total,
            counts['Individual'] / total
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])

        shadow_overlay_settings = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2,
            "pulse_speed": 0.1,
            "overlay_texture": 'static_noise' if shadow_theme != 'Unknown' else 'none',
            "animation_curve": 'linear_fluctuation',
            "audio_cue": 'faint_static' if shadow_theme != 'Unknown' else 'none'
        }
        focused_distortion = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "distortion_intensity": 0.7,
            "ripple_frequency": 0.5,
            "distortion_waveform": 'perlin_noise',
            "audio_cue_intensity": 0.5
        }
        whispering_glyphs = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "text": shadow_theme,
            "fade_duration": 3.0,
            "glyph_particle_effect": 'dissolving_particles',
            "glyph_animation_type": 'floating'
        }

        tone_of_voice = self.curriculum_output.get('recursive_journal_entry', {}).get('tone_of_voice', 'neutral')

        if tone_of_voice == 'encouraging':
            shadow_overlay_settings['desaturation_filter'] = max(0.0, shadow_overlay_settings['desaturation_filter'] - 0.1)
            focused_distortion['distortion_intensity'] = max(0.0, focused_distortion['distortion_intensity'] - 0.2)
            shadow_overlay_settings['audio_cue'] = 'subtle_chime'
            focused_distortion['audio_cue_intensity'] = max(0.0, focused_distortion['audio_cue_intensity'] - 0.2)
            whispering_glyphs['glyph_particle_effect'] = 'rising_sparkles'
            whispering_glyphs['glyph_animation_type'] = 'ascending'

        self.config['defrag_visuals']['shadow_overlay_settings'] = shadow_overlay_settings
        self.config['defrag_visuals']['focused_distortion'] = focused_distortion
        self.config['defrag_visuals']['whispering_glyphs'] = whispering_glyphs

        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            opportunity_visuals = {
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "particle_type": "sparkling_orbs", # New specific particle type
                "ephemeral_text": opp['message'],
                "channel_texture": 'flowing_energy',
                "channel_animation_curve": 'accelerating_flow',
                "audio_cue": 'chime'
            }

            if tone_of_voice == 'encouraging':
                opportunity_visuals['glow_intensity'] += 0.5 # Increase glow
                opportunity_visuals['glow_color'] = "#A3D4FF" # Brighter blue
                opportunity_visuals['channel_animation_curve'] = 'soaring_arc'
                opportunity_visuals['audio_cue'] = 'uplifting_swell'

            self.config['defrag_visuals']['opportunities'].append(opportunity_visuals)

    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        # Determine if the primary user (self.user) is pressured for influencing shared visuals
        user_is_pressured = self._is_user_pressured()

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_name": spark['source_1'],
                "source_2_name": spark['source_2'],
                "nature": spark['nature']
            }

            # Default values for new parameters
            geometry_deformation_type = 'none'
            geometry_deformation_intensity = 0.0
            particle_stream_direction = 'none'
            lighting_effect_type = 'none'
            environmental_aura_color = '#FFFFFF'
            audio_cue = 'none'

            if spark['nature'] == "Harmonious":
                geometry_deformation_type = 'subtle_pulse'
                geometry_deformation_intensity = 0.1
                particle_stream_direction = 'outward'
                lighting_effect_type = 'ambient_glow'
                environmental_aura_color = '#E0FFFF'
                audio_cue = 'harmonious_chime'
                visual_props.update({"color": "#ADD8E6", "animation_speed": 0.8, "particle_effect_type": "shimmer", "glow_intensity": 1.2})
            elif spark['nature'] == "Volatile/Energizing":
                geometry_deformation_type = 'erratic_shimmer'
                geometry_deformation_intensity = 0.4
                particle_stream_direction = 'random_flow'
                lighting_effect_type = 'focused_beam'
                environmental_aura_color = '#FFD700'
                audio_cue = 'static_hiss'
                visual_props.update({"color": "#FF4500", "animation_speed": 2.0, "particle_effect_type": "bursts", "glow_intensity": 1.8})
            elif spark['nature'] == "Restrictive/Binding":
                geometry_deformation_type = 'subtle_compression'
                geometry_deformation_intensity = 0.3
                particle_stream_direction = 'inward'
                lighting_effect_type = 'shadow_cast'
                environmental_aura_color = '#4B0082'
                audio_cue = 'deep_hum'
                visual_props.update({"color": "#6A5ACD", "animation_speed": 0.4, "particle_effect_type": "dense_mist", "glow_intensity": 0.6})

            # Apply influence from primary user's defrag output if they are source_1
            if user_is_pressured and spark['source_1_name'].startswith(self.user['name']):
                if 'Volatile' not in spark['nature']: # Only enhance non-volatile if user is pressured
                    geometry_deformation_intensity = min(1.0, geometry_deformation_intensity + 0.2)
                    particle_stream_direction = 'erratic_flicker'
                    lighting_effect_type = 'subtle_flicker'
                    audio_cue = 'underlying_tension'

            visual_props.update({
                "geometry_deformation_type": geometry_deformation_type,
                "geometry_deformation_intensity": geometry_deformation_intensity,
                "particle_stream_direction": particle_stream_direction,
                "lighting_effect_type": lighting_effect_type,
                "environmental_aura_color": environmental_aura_color,
                "audio_cue": audio_cue
            })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        # Mocking a friction point for demonstration if no actual friction points are generated
        if not self.relationship_output.get('friction_points'):
             # Example: create a mock friction point if source_1's sun gate is 38 (The Fighter - struggle)
            user_sun_gate = str(self.user['mechanics']['planetary_map'].get('Sun'))
            if user_sun_gate == '38' and hexagrammatic_kernel['gates'].get(user_sun_gate, {}).get('shadow') == 'Struggle':
                self.config['relationship_visuals']['friction_points'].append({
                    "type": "conflict",
                    "gates_involved": [38, hexagrammatic_kernel['harmonic_map'].get(38)], # Assuming a harmonic partner exists for potential conflict
                    "distortion_intensity": 0.9,
                    "color": "#FF0000",
                    "effect": "jagged_aura",
                    "geometry_deformation_type": 'fractured_surface',
                    "particle_stream_type": 'chaotic_bursts',
                    "lighting_effect_type": 'flickering_red',
                    "environmental_aura_color": '#FF8C00',
                    "audio_cue": 'dissonant_tone'
                })
        else:
            for friction in self.relationship_output.get('friction_points', []):
                # Existing friction points would get these parameters
                self.config['relationship_visuals']['friction_points'].append({
                    "type": "conflict",
                    "gates_involved": friction.get('gates', []),
                    "distortion_intensity": 0.9,
                    "color": "#FF0000",
                    "effect": "jagged_aura",
                    "geometry_deformation_type": 'fractured_surface',
                    "particle_stream_type": 'chaotic_bursts',
                    "lighting_effect_type": 'flickering_red',
                    "environmental_aura_color": '#FF8C00',
                    "audio_cue": 'dissonant_tone'
                })


    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow"
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            if '31-7' in core_focus:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Enhanced Relationship Visuals ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
du = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
ru = relationship_engine_instance.run_overlay()

# Manually inject a mock electromagnetic spark and a friction point for demonstration purposes if they are empty
# to ensure the CymaticRenderer's relationship visual logic is triggered.
if not ru['dynamics']:
    print("\n--- Manually injecting mock electromagnetic spark for demonstration ---")
    ru['dynamics'].append({
        "type": "ELECTROMAGNETIC_SPARK",
        "channel": "57-10", # Example channel between user_alpha's Saturn and user_beta's Earth
        "source_1": f"{user_alpha['name']}'s Saturn (57)",
        "source_2": f"{user_beta['name']}'s Earth (10)",
        "nature": "Restrictive/Binding",
        "description": "Restrictive/Binding connection between The Intuitive and The Self-Lover."
    })
    ru['dynamics'].append({
        "type": "ELECTROMAGNETIC_SPARK",
        "channel": "34-10", # Example channel between user_alpha's Pluto and user_beta's Earth
        "source_1": f"{user_alpha['name']}'s Pluto (34)",
        "source_2": f"{user_beta['name']}'s Earth (10)",
        "nature": "Harmonious",
        "description": "Harmonious connection between The Power Generator and The Self-Lover."
    })
if not ru['friction_points']:
    print("--- Manually injecting mock friction point for demonstration ---")
    # Example: a potential friction point if Alpha's sun gate implies struggle and it connects with something negative.
    # For this demo, let's just make a generic one to show the visuals are populated.
    ru['friction_points'].append({
        "type": "potential_conflict",
        "gates_involved": [31, 7], # Could be a channel, or just two gates that frequently clash
        "description": "Potential communication friction due to conflicting strategies."
    })

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
pu = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, du, tone_of_voice="encouraging")
cu = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=du,
    relationship_output=ru,
    penta_output=pu,
    curriculum_output=cu
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with enhanced relationship visuals. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {},
            "relationship_visuals": {},
            "penta_visuals": {},
            "curriculum_visuals": {}
        }

    def _is_user_pressured(self):
        # Helper to check if the main user (self.user) is under pressure from daily transits
        pressure_gates = [61, 63, 64]
        user_active_gates = self.user['mechanics']['active_gates'] # User's natal gates, not transit gates
        transit_active_gates = self.transits['mechanics']['active_gates'] # Active gates from today's transits

        # Check if the daily transits themselves are from a pressure gate
        if self.transits['mechanics']['sun_gate'] in pressure_gates:
            return True

        # Check if any of user's active gates are pressure gates and are part of the daily active transits (indirect pressure)
        if any(g in pressure_gates and g in transit_active_gates for g in user_active_gates):
            return True

        return False

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1

        vector = [
            counts['Collective'] / total,
            counts['Tribal'] / total,
            counts['Individual'] / total
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])

        shadow_overlay_settings = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2,
            "pulse_speed": 0.1,
            "overlay_texture": 'static_noise' if shadow_theme != 'Unknown' else 'none',
            "animation_curve": 'linear_fluctuation',
            "audio_cue": 'faint_static' if shadow_theme != 'Unknown' else 'none'
        }
        focused_distortion = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "distortion_intensity": 0.7,
            "ripple_frequency": 0.5,
            "distortion_waveform": 'perlin_noise',
            "audio_cue_intensity": 0.5
        }
        whispering_glyphs = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "text": shadow_theme,
            "fade_duration": 3.0,
            "glyph_particle_effect": 'dissolving_particles',
            "glyph_animation_type": 'floating'
        }

        tone_of_voice = self.curriculum_output.get('recursive_journal_entry', {}).get('tone_of_voice', 'neutral')

        if tone_of_voice == 'encouraging':
            shadow_overlay_settings['desaturation_filter'] = max(0.0, shadow_overlay_settings['desaturation_filter'] - 0.1)
            focused_distortion['distortion_intensity'] = max(0.0, focused_distortion['distortion_intensity'] - 0.2)
            shadow_overlay_settings['audio_cue'] = 'subtle_chime'
            focused_distortion['audio_cue_intensity'] = max(0.0, focused_distortion['audio_cue_intensity'] - 0.2)
            whispering_glyphs['glyph_particle_effect'] = 'rising_sparkles'
            whispering_glyphs['glyph_animation_type'] = 'ascending'

        self.config['defrag_visuals']['shadow_overlay_settings'] = shadow_overlay_settings
        self.config['defrag_visuals']['focused_distortion'] = focused_distortion
        self.config['defrag_visuals']['whispering_glyphs'] = whispering_glyphs

        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            opportunity_visuals = {
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "particle_type": "sparkling_orbs", # New specific particle type
                "ephemeral_text": opp['message'],
                "channel_texture": 'flowing_energy',
                "channel_animation_curve": 'accelerating_flow',
                "audio_cue": 'chime'
            }

            if tone_of_voice == 'encouraging':
                opportunity_visuals['glow_intensity'] += 0.5 # Increase glow
                opportunity_visuals['glow_color'] = "#A3D4FF" # Brighter blue
                opportunity_visuals['channel_animation_curve'] = 'soaring_arc'
                opportunity_visuals['audio_cue'] = 'uplifting_swell'

            self.config['defrag_visuals']['opportunities'].append(opportunity_visuals)

    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        # Determine if the primary user (self.user) is pressured for influencing shared visuals
        user_is_pressured = self._is_user_pressured()

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_label": spark['source_1'], # Use _label to store full descriptive string
                "source_2_label": spark['source_2'], # Use _label to store full descriptive string
                "nature": spark['nature']
            }

            # Default values for new parameters
            geometry_deformation_type = 'none'
            geometry_deformation_intensity = 0.0
            particle_stream_direction = 'none'
            lighting_effect_type = 'none'
            environmental_aura_color = '#FFFFFF'
            audio_cue = 'none'

            if spark['nature'] == "Harmonious":
                geometry_deformation_type = 'subtle_pulse'
                geometry_deformation_intensity = 0.1
                particle_stream_direction = 'outward'
                lighting_effect_type = 'ambient_glow'
                environmental_aura_color = '#E0FFFF'
                audio_cue = 'harmonious_chime'
                visual_props.update({"color": "#ADD8E6", "animation_speed": 0.8, "particle_effect_type": "shimmer", "glow_intensity": 1.2})
            elif spark['nature'] == "Volatile/Energizing":
                geometry_deformation_type = 'erratic_shimmer'
                geometry_deformation_intensity = 0.4
                particle_stream_direction = 'random_flow'
                lighting_effect_type = 'focused_beam'
                environmental_aura_color = '#FFD700'
                audio_cue = 'static_hiss'
                visual_props.update({"color": "#FF4500", "animation_speed": 2.0, "particle_effect_type": "bursts", "glow_intensity": 1.8})
            elif spark['nature'] == "Restrictive/Binding":
                geometry_deformation_type = 'subtle_compression'
                geometry_deformation_intensity = 0.3
                particle_stream_direction = 'inward'
                lighting_effect_type = 'shadow_cast'
                environmental_aura_color = '#4B0082'
                audio_cue = 'deep_hum'
                visual_props.update({"color": "#6A5ACD", "animation_speed": 0.4, "particle_effect_type": "dense_mist", "glow_intensity": 0.6})

            # Apply influence from primary user's defrag output if they are source_1
            # Corrected to use 'source_1' as it exists in the spark dictionary
            if user_is_pressured and spark['source_1'].startswith(self.user['name']):
                if 'Volatile' not in spark['nature']: # Only enhance non-volatile if user is pressured
                    geometry_deformation_intensity = min(1.0, geometry_deformation_intensity + 0.2)
                    particle_stream_direction = 'erratic_flicker'
                    lighting_effect_type = 'subtle_flicker'
                    audio_cue = 'underlying_tension'

            visual_props.update({
                "geometry_deformation_type": geometry_deformation_type,
                "geometry_deformation_intensity": geometry_deformation_intensity,
                "particle_stream_direction": particle_stream_direction,
                "lighting_effect_type": lighting_effect_type,
                "environmental_aura_color": environmental_aura_color,
                "audio_cue": audio_cue
            })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        # Mocking a friction point for demonstration if no actual friction points are generated
        if not self.relationship_output.get('friction_points'):
             # Example: create a mock friction point if source_1's sun gate is 38 (The Fighter - struggle)
            user_sun_gate = str(self.user['mechanics']['planetary_map'].get('Sun'))
            # Check if hexagrammatic_kernel['harmonic_map'] has the key 38 before accessing
            harmonic_partner_38 = hexagrammatic_kernel['harmonic_map'][38] if 38 in hexagrammatic_kernel['harmonic_map'] else None
            mock_gates_involved = [38] if 38 in hexagrammatic_kernel['gates'] else []
            if harmonic_partner_38 is not None:
                mock_gates_involved.append(harmonic_partner_38)

            if user_sun_gate == '38' and hexagrammatic_kernel['gates'].get(user_sun_gate, {}).get('shadow') == 'Struggle':
                self.config['relationship_visuals']['friction_points'].append({
                    "type": "conflict",
                    "gates_involved": mock_gates_involved,
                    "distortion_intensity": 0.9,
                    "color": "#FF0000",
                    "effect": "jagged_aura",
                    "geometry_deformation_type": 'fractured_surface',
                    "particle_stream_type": 'chaotic_bursts',
                    "lighting_effect_type": 'flickering_red',
                    "environmental_aura_color": '#FF8C00',
                    "audio_cue": 'dissonant_tone'
                })
        else:
            for friction in self.relationship_output.get('friction_points', []):
                # Existing friction points would get these parameters
                self.config['relationship_visuals']['friction_points'].append({
                    "type": "conflict",
                    "gates_involved": friction.get('gates_involved', []), # Use 'gates_involved' from input, not 'gates'
                    "distortion_intensity": 0.9,
                    "color": "#FF0000",
                    "effect": "jagged_aura",
                    "geometry_deformation_type": 'fractured_surface',
                    "particle_stream_type": 'chaotic_bursts',
                    "lighting_effect_type": 'flickering_red',
                    "environmental_aura_color": '#FF8C00',
                    "audio_cue": 'dissonant_tone'
                })


    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5
        }

        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', '')
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })

        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow"
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            if '31-7' in core_focus:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Enhanced Relationship Visuals ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
du = DefragEngine(user_alpha, daily_transits).run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
ru = relationship_engine_instance.run_overlay()

# Manually inject a mock electromagnetic spark and a friction point for demonstration purposes if they are empty
# to ensure the CymaticRenderer's relationship visual logic is triggered.
# Ensure injected sparks match the structure generated by RelationshipEngine (using 'source_1', 'source_2' etc.)
if not ru['dynamics']:
    print("\n--- Manually injecting mock electromagnetic spark for demonstration ---")
    ru['dynamics'].append({
        "type": "ELECTROMAGNETIC_SPARK",
        "channel": "57-10", # Example channel
        "source_1": f"{user_alpha['name']}'s Saturn (57)", # Full descriptive string
        "source_2": f"{user_beta['name']}'s Earth (10)",   # Full descriptive string
        "nature": "Restrictive/Binding",
        "description": "Restrictive/Binding connection between The Intuitive and The Self-Lover."
    })
    ru['dynamics'].append({
        "type": "ELECTROMAGNETIC_SPARK",
        "channel": "34-10", # Example channel
        "source_1": f"{user_alpha['name']}'s Pluto (34)",  # Full descriptive string
        "source_2": f"{user_beta['name']}'s Earth (10)",   # Full descriptive string
        "nature": "Harmonious",
        "description": "Harmonious connection between The Power Generator and The Self-Lover."
    })
if not ru['friction_points']:
    print("--- Manually injecting mock friction point for demonstration ---")
    # Example: a potential friction point if Alpha's sun gate implies struggle and it connects with something negative.
    # For this demo, let's just make a generic one to show the visuals are populated.
    ru['friction_points'].append({
        "type": "potential_conflict",
        "gates_involved": [31, 7], # Could be a channel, or just two gates that frequently clash
        "description": "Potential communication friction due to conflicting strategies."
    })

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
pu = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, du, tone_of_voice="encouraging")
cu = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=du,
    relationship_output=ru,
    penta_output=pu,
    curriculum_output=cu
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with enhanced relationship visuals. ---")
````

````
import json
import datetime
import random
from collections import Counter

# ==============================================================================
# SECTION 1: KNOWLEDGE BASE (THE "GOD MODE" DATASETS)
# ==============================================================================

# 1.1 THE HEXAGRAMMATIC KERNEL (64 GATES)
# ---------------------------------------------------
hexagrammatic_kernel = {
    "gates": {
        "1": {"label": "The Originator", "shadow": "Entropy", "gift": "Freshness", "circuit": "Individual"},
        "2": {"label": "The Compass", "shadow": "Dislocation", "gift": "Orientation", "circuit": "Individual"},
        "3": {"label": "The Innovator", "shadow": "Chaos", "gift": "Innovation", "circuit": "Individual"},
        "4": {"label": "The Logic Formulator", "shadow": "Intolerance", "gift": "Understanding", "circuit": "Collective"},
        "5": {"label": "The Rhythm Keeper", "shadow": "Impatience", "gift": "Patience", "circuit": "Collective"},
        "6": {"label": "The Peacemaker", "shadow": "Conflict", "gift": "Diplomacy", "circuit": "Tribal"},
        "7": {"label": "The Strategist", "shadow": "Division", "gift": "Guidance", "circuit": "Collective"},
        "8": {"label": "The Publicist", "shadow": "Mediocrity", "gift": "Style", "circuit": "Individual"},
        "9": {"label": "The Focus Engine", "shadow": "Inertia", "gift": "Determination", "circuit": "Collective"},
        "10": {"label": "The Self-Lover", "shadow": "Self-Obsession", "gift": "Naturalness", "circuit": "Individual"},
        "11": {"label": "The Conceptualist", "shadow": "Obscurity", "gift": "Idealism", "circuit": "Collective"},
        "12": {"label": "The Articulator", "shadow": "Vanity", "gift": "Discrimination", "circuit": "Individual"},
        "13": {"label": "The Witness", "shadow": "Discord", "gift": "Listening", "circuit": "Collective"},
        "14": {"label": "The Power House", "shadow": "Compromise", "gift": "Competence", "circuit": "Individual"},
        "15": {"label": "The Extremist", "shadow": "Dullness", "gift": "Magnetism", "circuit": "Collective"},
        "16": {"label": "The Skills Master", "shadow": "Indifference", "gift": "Versatility", "circuit": "Collective"},
        "17": {"label": "The Opinionator", "shadow": "Opinionated", "gift": "Far-Sightedness", "circuit": "Collective"},
        "18": {"label": "The Editor", "shadow": "Judgment", "gift": "Integrity", "circuit": "Collective"},
        "19": {"label": "The Attuner", "shadow": "Co-Dependence", "gift": "Sensitivity", "circuit": "Tribal"},
        "20": {"label": "The Now Awareness", "shadow": "Superficiality", "gift": "Self-Assurance", "circuit": "Individual"},
        "21": {"label": "The Treasurer", "shadow": "Control", "gift": "Authority", "circuit": "Tribal"},
        "22": {"label": "The Social Charmer", "shadow": "Dishonor", "gift": "Graciousness", "circuit": "Individual"},
        "23": {"label": "The Assimilator", "shadow": "Complexity", "gift": "Simplicity", "circuit": "Individual"},
        "24": {"label": "The Rationalizer", "shadow": "Addiction", "gift": "Invention", "circuit": "Individual"},
        "25": {"label": "The Spirit Warrior", "shadow": "Constriction", "gift": "Acceptance", "circuit": "Individual"},
        "26": {"label": "The Deal Maker", "shadow": "Pride", "gift": "Artfulness", "circuit": "Tribal"},
        "27": {"label": "The Caretaker", "shadow": "Selfishness", "gift": "Altruism", "circuit": "Tribal"},
        "28": {"label": "The Risk Taker", "shadow": "Purposelessness", "gift": "Totality", "circuit": "Individual"},
        "29": {"label": "The Committer", "shadow": "Half-Heartedness", "gift": "Commitment", "circuit": "Collective"},
        "30": {"label": "The Passionate Fire", "shadow": "Desire", "gift": "Lightness", "circuit": "Collective"},
        "31": {"label": "The Voice of Influence", "shadow": "Arrogance", "gift": "Leadership", "circuit": "Collective"},
        "32": {"label": "The Conservator", "shadow": "Failure", "gift": "Preservation", "circuit": "Tribal"},
        "33": {"label": "The Storyteller", "shadow": "Forgetting", "gift": "Mindfulness", "circuit": "Collective"},
        "34": {"label": "The Power Generator", "shadow": "Force", "gift": "Strength", "circuit": "Individual"},
        "35": {"label": "The Progress Driver", "shadow": "Hunger", "gift": "Adventure", "circuit": "Collective"},
        "36": {"label": "The Crisis Manager", "shadow": "Turbulence", "gift": "Humanity", "circuit": "Collective"},
        "37": {"label": "The Family Guardian", "shadow": "Weakness", "gift": "Equality", "circuit": "Tribal"},
        "38": {"label": "The Fighter", "shadow": "Struggle", "gift": "Perseverance", "circuit": "Individual"},
        "39": {"label": "The Provocateur", "shadow": "Provocation", "gift": "Dynamism", "circuit": "Individual"},
        "40": {"label": "The Deliverer", "shadow": "Exhaustion", "gift": "Resolve", "circuit": "Tribal"},
        "41": {"label": "The Visionary", "shadow": "Fantasy", "gift": "Anticipation", "circuit": "Collective"},
        "42": {"label": "The Finisher", "shadow": "Expectation", "gift": "Detachment", "circuit": "Collective"},
        "43": {"label": "The Insight Formulator", "shadow": "Deafness", "gift": "Insight", "circuit": "Individual"},
        "44": {"label": "The Pattern Master", "shadow": "Interference", "gift": "Teamwork", "circuit": "Tribal"},
        "45": {"label": "The King/Queen", "shadow": "Dominance", "gift": "Synergy", "circuit": "Tribal"},
        "46": {"label": "The Vessel Lover", "shadow": "Seriousness", "gift": "Delight", "circuit": "Collective"},
        "47": {"label": "The Alchemist", "shadow": "Oppression", "gift": "Transmutation", "circuit": "Collective"},
        "48": {"label": "The Deep Well", "shadow": "Inadequacy", "gift": "Resourcefulness", "circuit": "Collective"},
        "49": {"label": "The Revolutionary", "shadow": "Reaction", "gift": "Revolution", "circuit": "Tribal"},
        "50": {"label": "The Law Maker", "shadow": "Corruption", "gift": "Equilibrium", "circuit": "Tribal"},
        "51": {"label": "The Shock Master", "shadow": "Agitation", "gift": "Initiative", "circuit": "Individual"},
        "52": {"label": "The Mountain", "shadow": "Stress", "gift": "Restraint", "circuit": "Collective"},
        "53": {"label": "The Starter", "shadow": "Immaturity", "gift": "Expansion", "circuit": "Collective"},
        "54": {"label": "The Ambitious", "shadow": "Greed", "gift": "Aspiration", "circuit": "Tribal"},
        "55": {"label": "The Spirit Cup", "shadow": "Victimization", "gift": "Freedom", "circuit": "Individual"},
        "56": {"label": "The Wanderer", "shadow": "Distraction", "gift": "Enrichment", "circuit": "Collective"},
        "57": {"label": "The Intuitive", "shadow": "Unease", "gift": "Intuition", "circuit": "Individual"},
        "58": {"label": "The Joyous", "shadow": "Dissatisfaction", "gift": "Vitality", "circuit": "Collective"},
        "59": {"label": "The Bonder", "shadow": "Dishonesty", "gift": "Intimacy", "circuit": "Tribal"},
        "60": {"label": "The Realist", "shadow": "Limitation", "gift": "Realism", "circuit": "Individual"},
        "61": {"label": "The Mystery Processor", "shadow": "Psychosis", "gift": "Inspiration", "circuit": "Individual"},
        "62": {"label": "The Detailer", "shadow": "Intellect", "gift": "Precision", "circuit": "Collective"},
        "63": {"label": "The Doubter", "shadow": "Doubt", "gift": "Inquiry", "circuit": "Collective"},
        "64": {"label": "The Abstract Processor", "shadow": "Confusion", "gift": "Imagination", "circuit": "Collective"}
    }
}

# 1.2 THE PLANETARY ARCHETYPES
# ------------------------------------------------
planetary_archetypes = {
    "Sun": {"role": "Life's Work (70% Intensity)", "type": "Core"},
    "Earth": {"role": "Grounding & Evolution", "type": "Core"},
    "Moon": {"role": "Emotional Driver", "type": "Core"},
    "Mars": {"role": "Immature Energy & Conflict", "type": "Volatile"},
    "Jupiter": {"role": "Law & Protection", "type": "Benefic"},
    "Saturn": {"role": "The Judge & Restriction", "type": "Restrictive"},
    "Uranus": {"role": "Chaos & Innovation", "type": "Generational"},
    "Neptune": {"role": "Veil & Spirit", "type": "Generational"},
    "Pluto": {"role": "Truth & Transformation", "type": "Generational"}
}

# 1.3 THE RELATIONAL KERNEL (CHANNELS)
# --------------------------------------------------
channel_data = {
    "harmonic_map": {
        1:8, 8:1, 2:14, 14:2, 3:60, 60:3, 4:63, 63:4, 5:15, 15:5,
        6:59, 59:6, 7:31, 31:7, 9:52, 52:9, 10:20, 20:10, 10:34, 34:10,
        10:57, 57:10, 11:56, 56:11, 12:22, 22:12, 13:33, 33:13, 16:48, 48:16,
        17:62, 62:17, 18:58, 58:18, 19:49, 49:19, 20:34, 34:20, 20:57, 57:20,
        21:45, 45:21, 23:43, 43:23, 24:61, 61:24, 25:51, 51:25, 26:44, 44:26,
        27:50, 50:27, 28:38, 38:28, 29:46, 46:29, 30:41, 41:30, 32:54, 54:32,
        34:57, 57:34, 35:36, 36:35, 37:40, 40:37, 39:55, 55:39, 42:53, 53:42,
        47:64, 64:47
    }
}

# 1.4 THE PENTA KERNEL (GROUP DYNAMICS)
# -------------------------------------
penta_data = {
    "channels": {
        tuple(sorted((8, 1))): {"label": "Demonstration", "theme": "Leading by Example"},
        tuple(sorted((31, 7))): {"label": "Administration", "theme": "Planning & Strategy"},
        tuple(sorted((33, 13))): {"label": "Overseeing", "theme": "History & Accounting"},
        tuple(sorted((14, 2))): {"label": "Capacity", "theme": "Resources & Spin (The Backbone)"},
        tuple(sorted((15, 5))): {"label": "Flow", "theme": "Culture & Rhythm"},
        tuple(sorted((46, 29))): {"label": "Discovery", "theme": "Commitment & Work"}
    }
}

# 1.5 THE VISUAL KERNEL (CYMATIC MAPPING)
# ---------------------------------------
# Maps Energy Types and Circuits to Rendering properties
visual_kernel = {
    "circuitry_colors": {
        "Tribal": "#A52A2A",      # Red/Brown/Ochre (Earth/Blood)
        "Individual": "#00FF00",  # Neon Green/Purple (Mutation/Acoustic)
        "Collective": "#4682B4"   # Steel Blue/Grey (Logic/Patterns)
    },
    "type_waveforms": {
        "Generator": "Sine",             # Enveloping, rhythmic
        "Manifesting Generator": "Fast Sine", # Skipped beats
        "Projector": "Focused Beam",     # Penetrating, laser-like
        "Manifestor": "Sawtooth",        # Impacting, sharp
        "Reflector": "Flat/Sampling"     # Sampling, mirror-like
    },
    "center_frequencies": {
        "Head": 963, "Ajna": 852, "Throat": 741, "G": 639,
        "Heart": 528, "Solar Plexus": 417, "Sacral": 285,
        "Root": 174, "Spleen": 396
    }
}

# ==============================================================================
# SECTION 2: THE INTAKE ENGINE (PROFILE GENERATION)
# ==============================================================================

class ProfileGenerator:
    @staticmethod
    def create_user_profile(name, birth_date, birth_time, birth_city, role="generic"):
        is_tribal = role == "tribal_authority"
        is_flow = role == "flow_generator"

        # MOCK PLANETARY ACTIVATIONS
        if is_tribal:
            # Tribal Dad
            planets = {"Sun": 21, "Earth": 48, "Moon": 40, "Mars": 59, "Jupiter": 45, "Saturn": 50, "Pluto": 26}
            active_gates = list(planets.values())
        elif is_flow:
            # Flow Mom
            planets = {"Sun": 15, "Earth": 10, "Moon": 5, "Mars": 29, "Jupiter": 1, "Saturn": 8, "Pluto": 27}
            active_gates = list(planets.values())
        else:
            # Standard Alpha
            planets = {"Sun": 31, "Earth": 41, "Moon": 48, "Mars": 14, "Jupiter": 29, "Saturn": 57, "Pluto": 34}
            active_gates = list(planets.values())

        profile = {
            "user_id": f"USER_{random.randint(1000,9999)}_{role.upper()}",
            "name": name,
            "meta": {"birth_date": birth_date, "role": role},
            "mechanics": {
                "type": "Manifestor" if is_tribal else "Generator",
                "authority": "Emotional" if is_tribal else "Sacral",
                "profile": "2/4" if is_tribal else "5/1",
                "active_gates": active_gates,
                "planetary_map": planets
            }
        }
        return profile

# ==============================================================================
# SECTION 3: THE DEFRAG ENGINE (INDIVIDUAL & DAILY COMPUTE)
# ==============================================================================

class DefragEngine:
    def __init__(self, user_profile, daily_transits):
        self.user = user_profile
        self.transits = daily_transits
        self.readout = {
            "meta": {"mode": "INDIVIDUAL_DAILY_SCAN", "user": user_profile['name']},
            "alerts": [],
            "opportunities": [],
            "shadow_work": {},
            "render_config": {}
        }

    def run_daily_scan(self):
        self._analyze_connections()
        self._analyze_shadow()
        # Adding transits to readout to help CurriculumEngine get gift state for shadow theme
        self.readout['transits'] = self.transits # Store transits directly for easier access
        return self.readout

    def _analyze_connections(self):
        harmonic_map = channel_data['harmonic_map']
        for planet, u_gate in self.user['mechanics']['planetary_map'].items():
            if u_gate in harmonic_map:
                partner = harmonic_map[u_gate]
                if partner in self.transits['mechanics']['active_gates']:
                    my_gate = hexagrammatic_kernel['gates'][str(u_gate)]
                    transit_gate = hexagrammatic_kernel['gates'][str(partner)]
                    planet_role = planetary_archetypes.get(planet, {}).get("role", "Unknown")

                    self.readout['opportunities'].append({
                        "type": "TRANSIT_SUPERPOWER",
                        "channel": f"{u_gate}-{partner}",
                        "nuance": f"Activated by your {planet} ({planet_role})",
                        "message": f"Your {planet} in {u_gate} ({my_gate['label']}) connects with the Field's {partner} ({transit_gate['label']}).",
                        "advice": "Use this specific planetary energy to bridge the gap today."
                    })

    def _analyze_shadow(self):
        sun_gate = str(self.transits['mechanics']['sun_gate'])
        sun_info = hexagrammatic_kernel['gates'].get(sun_gate)
        if sun_info:
            self.readout['shadow_work'] = {
                "theme": sun_info['shadow'],
                "inquiry": f"Where is {sun_info['shadow']} governing your choices today?"
            }

# ==============================================================================
# SECTION 4: THE RELATIONSHIP ENGINE (OVERLAY)
# ==============================================================================

class RelationshipEngine:
    def __init__(self, user_a, user_b):
        self.person_1 = user_a
        self.person_2 = user_b
        self.readout = {
            "meta": {"mode": "RELATIONSHIP_OVERLAY", "subjects": [user_a['name'], user_b['name']]},
            "dynamics": [],
            "friction_points": []
        }

    def run_overlay(self):
        self._analyze_electromagnetics()
        return self.readout

    def _analyze_electromagnetics(self):
        harmonic_map = channel_data['harmonic_map']
        p1_map = self.person_1['mechanics']['planetary_map']
        p2_map = self.person_2['mechanics']['planetary_map']

        for planet_1, g1 in p1_map.items():
            if g1 in harmonic_map:
                partner = harmonic_map[g1]
                planet_2 = next((p for p, g in p2_map.items() if g == partner), None)

                if planet_2:
                    gate_1_info = hexagrammatic_kernel['gates'][str(g1)]
                    gate_2_info = hexagrammatic_kernel['gates'][str(partner)]
                    interaction_type = "Harmonious"
                    if "Mars" in [planet_1, planet_2]: interaction_type = "Volatile/Energizing"
                    if "Saturn" in [planet_1, planet_2]: interaction_type = "Restrictive/Binding"

                    self.readout['dynamics'].append({
                        "type": "ELECTROMAGNETIC_SPARK",
                        "channel": f"{g1}-{partner}",
                        "source_1": f"{self.person_1['name']}'s {planet_1} ({g1})",
                        "source_2": f"{self.person_2['name']}'s {planet_2} ({partner})",
                        "nature": interaction_type,
                        "description": f"{interaction_type} connection between {gate_1_info['label']} and {gate_2_info['label']}."
                    })

# ==============================================================================
# SECTION 5: THE PENTA ENGINE (FAMILY SYSTEM)
# ==============================================================================

class PentaEngine:
    def __init__(self, family_members):
        self.members = family_members
        self.readout = {
            "meta": {"mode": "FAMILY_PENTA_SCAN"},
            "functional_health": "",
            "circuitry_balance": {},
            "defined_penta_channels": [] # Added to store defined Penta channels
        }

    def run_penta_analysis(self):
        self._analyze_circuitry()
        self._analyze_penta_channels()
        return self.readout

    def _analyze_circuitry(self):
        counts = Counter()
        total_gates = 0
        for m in self.members:
            for gate in m['mechanics']['active_gates']:
                gate_info = hexagrammatic_kernel['gates'].get(str(gate))
                if gate_info:
                    counts[gate_info['circuit']] += 1
                    total_gates += 1

        balance = {k: f"{round((v/total_gates)*100)}%" for k, v in counts.items()}
        self.readout['circuitry_balance'] = balance
        dom = counts.most_common(1)[0][0]
        if dom == "Tribal": self.readout['circuitry_note'] = "High Tribal Energy. Values Loyalty."
        elif dom == "Individual": self.readout['circuitry_note'] = "High Individual Energy. Values Mutation/Freedom."
        elif dom == "Collective": self.readout['circuitry_note'] = "High Collective Energy. Values Logic/Sharing."

    def _analyze_penta_channels(self):
        all_active_gates = set()
        for m in self.members:
            all_active_gates.update(m['mechanics']['active_gates'])

        defined_count = 0
        for (g1, g2), info in penta_data['channels'].items():
            if g1 in all_active_gates and g2 in all_active_gates:
                defined_count += 1
                # Store defined Penta channels
                self.readout['defined_penta_channels'].append({
                    "channel_gates": sorted((g1, g2)),
                    "label": info["label"],
                    "theme": info["theme"]
                })

        if defined_count < 3: self.readout['functional_health'] = "DYSFUNCTIONAL: Missing structural glue."
        else: self.readout['functional_health'] = "FUNCTIONAL: Strong operational unit."

# ==============================================================================
# SECTION 6: THE CURRICULUM & JOURNAL ENGINE (STRATEGIC OUTPUT)
# ==============================================================================

class CurriculumEngine:
    """
    Generates the 'Recursive Journal' and 'Strategic Curriculum'.
    Input: Daily Scan Results + User Profile.
    Output: Actionable prompts and lesson plans.
    """
    def __init__(self, user_profile, daily_scan_result, tone_of_voice="neutral"):
        self.user = user_profile
        self.scan = daily_scan_result
        self.tone_of_voice = tone_of_voice # New parameter
        self.readout = {
            "meta": {"mode": "CURRICULUM_GENERATION"},
            "recursive_journal_entry": {},
            "lesson_module": {}
        }

    def generate_curriculum(self):
        self._generate_journal_prompt()
        self._generate_lesson_plan()
        return self.readout

    def _generate_journal_prompt(self):
        # LOGIC: Pull the main Shadow theme from the Daily Scan
        # and create a 'Recursive' prompt that references past behavior (simulated).
        shadow_data = self.scan.get('shadow_work', {})
        theme = shadow_data.get('theme', 'Unknown')
        # Ensure we can get the gift state if sun_gate exists in scan and kernel
        gift = hexagrammatic_kernel['gates'][str(self.scan['transits']['mechanics']['sun_gate'])]['gift'] if 'transits' in self.scan and 'mechanics' in self.scan['transits'] and 'sun_gate' in self.scan['transits']['mechanics'] and str(self.scan['transits']['mechanics']['sun_gate']) in hexagrammatic_kernel['gates'] else 'its Gift state'

        base_text = f"Today's field is defined by {theme}. In your past logs, this frequency often triggers [Simulation: Anxiety/Inaction]. How can you transmute {theme} into {gift} today?"

        toned_text = base_text
        if self.tone_of_voice == "encouraging":
            toned_text = f"Remember, you have the strength to overcome anything. Today's theme is {theme}. Historically, this might have felt challenging. How can you bravely transform {theme} into its potent gift of {gift} today? You got this!"
        elif self.tone_of_voice == "scientific":
            toned_text = f"Analysis indicates the dominant transit theme for today is {theme}. Prior empirical observations suggest this frequency correlates with [Simulation: Anxiety/Inaction]. Formulate a strategy to re-contextualize {theme} towards its manifest gift of {gift}."
        # Add more tones as needed

        self.readout['recursive_journal_entry'] = {
            "date": datetime.date.today().isoformat(),
            "theme": theme,
            "prompt_type": "RECURSIVE_REFLECTION",
            "tone_of_voice": self.tone_of_voice,
            "text": toned_text,
            "tags": [theme, "Shadow Work", "Daily Log", self.tone_of_voice]
        }

    def _generate_lesson_plan(self):
        # LOGIC: Combine Profile (e.g., 5/1) with Planetary Opportunities
        profile = self.user['mechanics']['profile']
        opps = self.scan.get('opportunities', [])

        focus = "General Maintenance"
        if opps:
            focus = f"Leveraging {opps[0]['channel']} ({opps[0]['nuance']})"

        self.readout['lesson_module'] = {
            "title": f"Strategic Alignment for {profile} Profile",
            "core_focus": focus,
            "action_step": "Review the 'Transit Superpower' identified in your scan and apply it to one business decision today."
        }

# ==============================================================================
# SECTION 7: THE CYMATIC ENGINE (VECTOR CUBE & VISUALIZATION)
# ==============================================================================

class CymaticRenderer:
    """
    Translates Psychological Data into Physics/Visual Data.
    Output: A JSON config for a Three.js / p5.js frontend (The Vector Cube).
    """
    def __init__(self, user_profile, daily_transits, defrag_output=None, relationship_output=None, penta_output=None, curriculum_output=None):
        self.user = user_profile
        self.transits = daily_transits
        self.defrag_output = defrag_output
        self.relationship_output = relationship_output
        self.penta_output = penta_output
        self.curriculum_output = curriculum_output
        self.config = {
            "mandala_settings": {},
            "vector_slide_cube": {},
            "animation_states": [],
            "defrag_visuals": {},
            "relationship_visuals": {},
            "penta_visuals": {},
            "curriculum_visuals": {}
        }

    def _is_user_pressured(self):
        # Helper to check if the main user (self.user) is under pressure from daily transits
        pressure_gates = [61, 63, 64]
        user_active_gates = self.user['mechanics']['active_gates'] # User's natal gates, not transit gates
        transit_active_gates = self.transits['mechanics']['active_gates'] # Active gates from today's transits

        # Check if the daily transits themselves are from a pressure gate
        if self.transits['mechanics']['sun_gate'] in pressure_gates:
            return True

        # Check if any of user's active gates are pressure gates and are part of the daily active transits (indirect pressure)
        if any(g in pressure_gates and g in transit_active_gates for g in user_active_gates):
            return True

        return False

    def generate_config(self):
        self._set_base_geometry()
        self._set_vector_cube()
        self._set_color_palette()
        self._calculate_frequencies()
        self._set_animation_state()
        self._set_defrag_visuals()
        self._set_relationship_visuals()
        self._set_penta_visuals()
        self._set_curriculum_visuals()
        return self.config

    def _set_base_geometry(self):
        user_type = self.user['mechanics']['type']
        waveform = visual_kernel['type_waveforms'].get(user_type, "Sine")

        self.config['mandala_settings']['base_geometry'] = {
            "waveform": waveform,
            "geometry_complexity": len(self.user['mechanics']['active_gates']),
            "profile_structure": self.user['mechanics']['profile']
        }

    def _set_vector_cube(self):
        counts = Counter()
        total = 0
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info:
                counts[info['circuit']] += 1
                total += 1

        if total == 0: total = 1

        vector = [
            counts['Collective'] / total,
            counts['Tribal'] / total,
            counts['Individual'] / total
        ]

        self.config['vector_slide_cube'] = {
            "vector_coordinates": vector,
            "cube_rotation_speed": 0.05,
            "vertex_displacement": 0.2 if vector[2] > 0.5 else 0.05
        }

    def _set_color_palette(self):
        counts = Counter()
        for gate in self.user['mechanics']['active_gates']:
            info = hexagrammatic_kernel['gates'].get(str(gate))
            if info: counts[info['circuit']] += 1

        dominant_circuit = counts.most_common(1)[0][0]
        primary_hex = visual_kernel['circuitry_colors'].get(dominant_circuit, "#FFFFFF")

        self.config['mandala_settings']['color_palette'] = {
            "dominant_circuit": dominant_circuit,
            "primary_color": primary_hex,
            "secondary_color": "#FFFFFF",
            "glow_intensity": 0.8
        }

    def _calculate_frequencies(self):
        frequencies = []
        if self.user['mechanics']['type'] == "Manifestor":
            frequencies.append(visual_kernel['center_frequencies']['Throat'])
        elif self.user['mechanics']['type'] == "Generator":
            frequencies.append(visual_kernel['center_frequencies']['Sacral'])
        self.config['mandala_settings']['harmonic_frequencies_hz'] = frequencies

    def _set_animation_state(self):
        pressure_gates = [61, 63, 64]
        active_pressure = [g for g in pressure_gates if g in self.transits['mechanics']['active_gates']]

        if active_pressure:
            self.config['animation_states'].append({
                "state": "HIGH_PRESSURE_JITTER",
                "speed_multiplier": 2.5,
                "noise_level": 0.8
            })
        else:
            self.config['animation_states'].append({
                "state": "FLOW_STATE_BREATHING",
                "speed_multiplier": 1.0,
                "noise_level": 0.1
            })

    def _set_defrag_visuals(self):
        if not self.defrag_output:
            return

        shadow_theme = self.defrag_output['shadow_work'].get('theme', 'Unknown')
        shadow_gate_str = str(self.transits['mechanics']['sun_gate'])

        shadow_overlay_settings = {
            "active": True if shadow_theme != 'Unknown' else False,
            "theme": shadow_theme,
            "desaturation_filter": 0.2,
            "pulse_speed": 0.1,
            "overlay_texture": 'static_noise' if shadow_theme != 'Unknown' else 'none',
            "animation_curve": 'linear_fluctuation',
            "audio_cue": 'faint_static' if shadow_theme != 'Unknown' else 'none'
        }
        focused_distortion = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "distortion_intensity": 0.7,
            "ripple_frequency": 0.5,
            "distortion_waveform": 'perlin_noise',
            "audio_cue_intensity": 0.5
        }
        whispering_glyphs = {
            "active_gate": int(shadow_gate_str) if shadow_gate_str.isdigit() else None,
            "text": shadow_theme,
            "fade_duration": 3.0,
            "glyph_particle_effect": 'dissolving_particles',
            "glyph_animation_type": 'floating'
        }

        tone_of_voice = self.curriculum_output.get('recursive_journal_entry', {}).get('tone_of_voice', 'neutral')

        if tone_of_voice == 'encouraging':
            shadow_overlay_settings['desaturation_filter'] = max(0.0, shadow_overlay_settings['desaturation_filter'] - 0.1)
            focused_distortion['distortion_intensity'] = max(0.0, focused_distortion['distortion_intensity'] - 0.2)
            shadow_overlay_settings['audio_cue'] = 'subtle_chime'
            focused_distortion['audio_cue_intensity'] = max(0.0, focused_distortion['audio_cue_intensity'] - 0.2)
            whispering_glyphs['glyph_particle_effect'] = 'rising_sparkles'
            whispering_glyphs['glyph_animation_type'] = 'ascending'

        self.config['defrag_visuals']['shadow_overlay_settings'] = shadow_overlay_settings
        self.config['defrag_visuals']['focused_distortion'] = focused_distortion
        self.config['defrag_visuals']['whispering_glyphs'] = whispering_glyphs

        self.config['defrag_visuals']['opportunities'] = []
        for opp in self.defrag_output.get('opportunities', []):
            channel_parts = [int(p) for p in opp['channel'].split('-')]
            opportunity_visuals = {
                "type": opp['type'],
                "channel": channel_parts,
                "glow_color": "#FFD700", # Gold for superpower
                "glow_intensity": 1.5,
                "orbital_particles": True,
                "particle_type": "sparkling_orbs", # New specific particle type
                "ephemeral_text": opp['message'],
                "channel_texture": 'flowing_energy',
                "channel_animation_curve": 'accelerating_flow',
                "audio_cue": 'chime'
            }

            if tone_of_voice == 'encouraging':
                opportunity_visuals['glow_intensity'] += 0.5 # Increase glow
                opportunity_visuals['glow_color'] = "#A3D4FF" # Brighter blue
                opportunity_visuals['channel_animation_curve'] = 'soaring_arc'
                opportunity_visuals['audio_cue'] = 'uplifting_swell'

            self.config['defrag_visuals']['opportunities'].append(opportunity_visuals)

    def _set_relationship_visuals(self):
        if not self.relationship_output:
            return

        user_is_pressured = self._is_user_pressured()

        self.config['relationship_visuals']['electromagnetic_sparks'] = []
        for spark in self.relationship_output.get('dynamics', []):
            visual_props = {
                "channel": [int(p) for p in spark['channel'].split('-')],
                "source_1_label": spark['source_1'],
                "source_2_label": spark['source_2'],
                "nature": spark['nature']
            }

            geometry_deformation_type = 'none'
            geometry_deformation_intensity = 0.0
            particle_stream_direction = 'none'
            lighting_effect_type = 'none'
            environmental_aura_color = '#FFFFFF'
            audio_cue = 'none'

            if spark['nature'] == "Harmonious":
                geometry_deformation_type = 'subtle_pulse'
                geometry_deformation_intensity = 0.1
                particle_stream_direction = 'outward'
                lighting_effect_type = 'ambient_glow'
                environmental_aura_color = '#E0FFFF'
                audio_cue = 'harmonious_chime'
                visual_props.update({"color": "#ADD8E6", "animation_speed": 0.8, "particle_effect_type": "shimmer", "glow_intensity": 1.2})
            elif spark['nature'] == "Volatile/Energizing":
                geometry_deformation_type = 'erratic_shimmer'
                geometry_deformation_intensity = 0.4
                particle_stream_direction = 'random_flow'
                lighting_effect_type = 'focused_beam'
                environmental_aura_color = '#FFD700'
                audio_cue = 'static_hiss'
                visual_props.update({"color": "#FF4500", "animation_speed": 2.0, "particle_effect_type": "bursts", "glow_intensity": 1.8})
            elif spark['nature'] == "Restrictive/Binding":
                geometry_deformation_type = 'subtle_compression'
                geometry_deformation_intensity = 0.3
                particle_stream_direction = 'inward'
                lighting_effect_type = 'shadow_cast'
                environmental_aura_color = '#4B0082'
                audio_cue = 'deep_hum'
                visual_props.update({"color": "#6A5ACD", "animation_speed": 0.4, "particle_effect_type": "dense_mist", "glow_intensity": 0.6})

            if user_is_pressured and spark['source_1'].startswith(self.user['name']):
                if 'Volatile' not in spark['nature']:
                    geometry_deformation_intensity = min(1.0, geometry_deformation_intensity + 0.2)
                    particle_stream_direction = 'erratic_flicker'
                    lighting_effect_type = 'subtle_flicker'
                    audio_cue = 'underlying_tension'

            visual_props.update({
                "geometry_deformation_type": geometry_deformation_type,
                "geometry_deformation_intensity": geometry_deformation_intensity,
                "particle_stream_direction": particle_stream_direction,
                "lighting_effect_type": lighting_effect_type,
                "environmental_aura_color": environmental_aura_color,
                "audio_cue": audio_cue
            })
            self.config['relationship_visuals']['electromagnetic_sparks'].append(visual_props)

        self.config['relationship_visuals']['friction_points'] = []
        if not self.relationship_output.get('friction_points'):
            user_sun_gate = str(self.user['mechanics']['planetary_map'].get('Sun'))
            mock_gates_involved = []
            if user_sun_gate.isdigit() and int(user_sun_gate) in hexagrammatic_kernel['harmonic_map']:
                mock_gates_involved = [int(user_sun_gate), hexagrammatic_kernel['harmonic_map'][int(user_sun_gate)]]
            elif user_sun_gate.isdigit():
                mock_gates_involved = [int(user_sun_gate)]

            if user_sun_gate == '38' and hexagrammatic_kernel['gates'].get(user_sun_gate, {}).get('shadow') == 'Struggle':
                self.config['relationship_visuals']['friction_points'].append({
                    "type": "conflict",
                    "gates_involved": mock_gates_involved,
                    "distortion_intensity": 0.9,
                    "color": "#FF0000",
                    "effect": "jagged_aura",
                    "geometry_deformation_type": 'fractured_surface',
                    "particle_stream_type": 'chaotic_bursts',
                    "lighting_effect_type": 'flickering_red',
                    "environmental_aura_color": '#FF8C00',
                    "audio_cue": 'dissonant_tone'
                })
        else:
            for friction in self.relationship_output.get('friction_points', []):
                self.config['relationship_visuals']['friction_points'].append({
                    "type": "conflict",
                    "gates_involved": friction.get('gates_involved', friction.get('gates', [])),
                    "distortion_intensity": 0.9,
                    "color": "#FF0000",
                    "effect": "jagged_aura",
                    "geometry_deformation_type": 'fractured_surface',
                    "particle_stream_type": 'chaotic_bursts',
                    "lighting_effect_type": 'flickering_red',
                    "environmental_aura_color": '#FF8C00',
                    "audio_cue": 'dissonant_tone'
                })

    def _set_penta_visuals(self):
        if not self.penta_output:
            return

        # Functional Health Visualization
        functional_health = self.penta_output.get('functional_health', 'UNKNOWN')
        self.config['penta_visuals']['penta_structure_integrity'] = {
            "status": functional_health,
            "main_color": "#2ECC71" if "FUNCTIONAL" in functional_health else "#E74C3C",
            "animation_style": "pulsing_steady" if "FUNCTIONAL" in functional_health else "fragmented_flicker",
            "effect_intensity": 1.0 if "FUNCTIONAL" in functional_health else 1.5,
            "geometry_type": 'solid_mesh' if "FUNCTIONAL" in functional_health else 'fragmented_cloud',
            "vibrancy_level": 1.0 if "FUNCTIONAL" in functional_health else 0.5,
            "cohesion_animation": 'smooth_pulse' if "FUNCTIONAL" in functional_health else 'erratic_flicker',
            "audio_health_cue": 'harmonious_drone' if "FUNCTIONAL" in functional_health else 'dissonant_hum'
        }

        # Circuitry Balance Visualization
        circuitry_balance = self.penta_output.get('circuitry_balance', {})
        self.config['penta_visuals']['circuitry_balance_visuals'] = {
            "data": [],
            "dominant_circuit_note": self.penta_output.get('circuitry_note', ''),
            "display_type": 'segmented_ring',
            "color_scheme": {},
            "animation_type": 'smooth_fill',
            "audio_balance_cue": 'gentle_sweep'
        }
        for circuit, percentage_str in circuitry_balance.items():
            self.config['penta_visuals']['circuitry_balance_visuals']['data'].append({
                "circuit_type": circuit,
                "percentage": float(percentage_str.strip('%')),
                "color": visual_kernel['circuitry_colors'].get(circuit, "#CCCCCC")
            })
        # Populate color_scheme explicitly
        self.config['penta_visuals']['circuitry_balance_visuals']['color_scheme'] = visual_kernel['circuitry_colors']

        # Key Penta Channels Visualization
        defined_penta_channels = self.penta_output.get('defined_penta_channels', [])
        self.config['penta_visuals']['penta_channel_animations'] = []
        for channel_info in defined_penta_channels:
            self.config['penta_visuals']['penta_channel_animations'].append({
                "channel_gates": channel_info['channel_gates'],
                "label": channel_info['label'],
                "theme": channel_info['theme'],
                "animation_properties": {
                    "glow_strength": 1.8,
                    "flow_speed": 1.2,
                    "color": "#FFFF00",
                    "particle_type": "energy_flow",
                    "channel_texture_type": 'flowing_plasma',
                    "movement_indicator_type": 'directional_arrows',
                    "animation_curve": 'ease_in_out',
                    "audio_channel_cue": 'subtle_chime_whoosh'
                }
            })

    def _set_curriculum_visuals(self):
        if not self.curriculum_output:
            return

        journal_entry = self.curriculum_output.get('recursive_journal_entry', {})
        lesson_module = self.curriculum_output.get('lesson_module', {})

        self.config['curriculum_visuals']['journal_entry_visuals'] = {}
        if journal_entry:
            theme = journal_entry.get('theme', 'Unknown')
            tone_of_voice = journal_entry.get('tone_of_voice', 'neutral')

            theme_visuals = {
                'Doubt': {'background_texture': 'swirling_haze', 'animated_motif': 'question_marks', 'typography_style': 'jagged_font_shadow'},
                'Chaos': {'background_texture': 'fragmented_shapes', 'animated_motif': 'dynamic_geometric', 'typography_style': 'sharp_fractal_font'},
                'Intolerance': {'background_texture': 'harsh_lines', 'animated_motif': 'clashing_symbols', 'typography_style': 'rigid_block_font'},
                'Unknown': {'background_texture': 'subtle_noise', 'animated_motif': 'fading_dots', 'typography_style': 'standard_serif'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(theme_visuals.get(theme, theme_visuals['Unknown']))

            tone_visuals = {
                'encouraging': {'aesthetic_filter': 'warm_soft_lighting', 'border_style': 'glowing_border', 'color_palette_shift': 'boost_saturation'},
                'scientific': {'aesthetic_filter': 'minimalist_grid', 'border_style': 'precise_frame', 'color_palette_shift': 'cooler_analytical'},
                'neutral': {'aesthetic_filter': 'balanced_understated', 'border_style': 'simple_line', 'color_palette_shift': 'none'}
            }
            self.config['curriculum_visuals']['journal_entry_visuals'].update(tone_visuals.get(tone_of_voice, tone_visuals['neutral']))

        self.config['curriculum_visuals']['lesson_module_visuals'] = {}
        if lesson_module:
            core_focus = lesson_module.get('core_focus', '')
            action_step_text = lesson_module.get('action_step', '')

            if '31-7' in core_focus:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'speech_bubble_compass'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = True
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = True
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['thematic_icon'] = 'general_focus_icon'
                self.config['curriculum_visuals']['lesson_module_visuals']['mini_circuitry_animation_active'] = False
                self.config['curriculum_visuals']['lesson_module_visuals']['animated_keywords_active'] = False

            self.config['curriculum_visuals']['lesson_module_visuals']['progress_indicator_type'] = 'progress_bar'
            if 'business decision' in action_step_text:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'gear_icon'
            else:
                self.config['curriculum_visuals']['lesson_module_visuals']['action_icon'] = 'checkmark_icon'
            self.config['curriculum_visuals']['lesson_module_visuals']['gamified_elements_active'] = True


# ==============================================================================
# SECTION 8: THE SIMULATION (FULL STACK)
# ==============================================================================

print("\n--- Running Full Simulation with Enhanced PentaEngine Visuals ---")

# 1. GENERATE PROFILES (Re-run to ensure all variables are fresh)
user_alpha = ProfileGenerator.create_user_profile("Subject Alpha", "1990-01-01", "12:00", "City", role="standard")
user_beta = ProfileGenerator.create_user_profile("Subject Beta", "1992-05-15", "08:30", "Another City", role="flow_generator")
user_gamma = ProfileGenerator.create_user_profile("Subject Gamma", "1985-03-20", "22:15", "Another Town", role="tribal_authority")
user_delta = ProfileGenerator.create_user_profile("Subject Delta", "1995-11-05", "04:45", "Faraway City", role="standard")

# 2. SIMULATE DAILY TRANSITS
daily_transits = {"mechanics": {"sun_gate": 63, "active_gates": [63, 7, 4]}}

# 3. RUN DEFRAG ENGINE
defrag_engine_instance = DefragEngine(user_alpha, daily_transits)
du = defrag_engine_instance.run_daily_scan()

# 4. RUN RELATIONSHIP ENGINE
relationship_engine_instance = RelationshipEngine(user_alpha, user_beta)
ru = relationship_engine_instance.run_overlay()

# Manually inject a mock electromagnetic spark and a friction point for demonstration purposes if they are empty
# to ensure the CymaticRenderer's relationship visual logic is triggered.
if not ru['dynamics']:
    print("\n--- Manually injecting mock electromagnetic spark for demonstration ---")
    ru['dynamics'].append({
        "type": "ELECTROMAGNETIC_SPARK",
        "channel": "57-10", # Example channel between user_alpha's Saturn and user_beta's Earth
        "source_1": f"{user_alpha['name']}'s Saturn (57)",
        "source_2": f"{user_beta['name']}'s Earth (10)",
        "nature": "Restrictive/Binding",
        "description": "Restrictive/Binding connection between The Intuitive and The Self-Lover."
    })
    ru['dynamics'].append({
        "type": "ELECTROMAGNETIC_SPARK",
        "channel": "34-10", # Example channel between user_alpha's Pluto and user_beta's Earth
        "source_1": f"{user_alpha['name']}'s Pluto (34)",
        "source_2": f"{user_beta['name']}'s Earth (10)",
        "nature": "Harmonious",
        "description": "Harmonious connection between The Power Generator and The Self-Lover."
    })
if not ru['friction_points']:
    print("--- Manually injecting mock friction point for demonstration ---")
    ru['friction_points'].append({
        "type": "potential_conflict",
        "gates_involved": [31, 7],
        "description": "Potential communication friction due to conflicting strategies."
    })

# 5. RUN PENTA ENGINE
family_members = [user_alpha, user_beta, user_gamma, user_delta]
penta_engine_instance = PentaEngine(family_members)
pu = penta_engine_instance.run_penta_analysis()

# 6. RUN CURRICULUM ENGINE (with an arbitrary tone for this test)
curriculum_engine_instance = CurriculumEngine(user_alpha, du, tone_of_voice="encouraging")
cu = curriculum_engine_instance.generate_curriculum()

# 7. RUN CYMATIC RENDERER with all outputs
cymatic_renderer_instance = CymaticRenderer(
    user_profile=user_alpha,
    daily_transits=daily_transits,
    defrag_output=du,
    relationship_output=ru,
    penta_output=pu,
    curriculum_output=cu
)

final_cymatic_config = cymatic_renderer_instance.generate_config()
print(json.dumps(final_cymatic_config, indent=2))

print("\n--- CymaticRenderer config generated with enhanced PentaEngine visuals. ---")
````
